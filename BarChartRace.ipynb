```python
import pandas as pd
import bar_chart_race as bcr
import numpy as np
```

```python
df = pd.read_csv('full_grouped.csv', index_col='Date')
```

```python
df.head(10)
```

```python
bcr_df = df['2020-06-01':'2020-07-01'].pivot_table(values = ['New cases'], 
                        index =['Date'], 
                        columns = ['WHO Region'], 
                        aggfunc = np.sum )

bcr_df.columns = ['África','Américas', 
                  'Mediterrâneo Oriental',
                  'Europa',
                  'Sudeste Asiático', 
                  'Pacífico Oeste']
```

```python
bcr_df.head(10)
```

```python
bcr.bar_chart_race(df = bcr_df)
```

```python
bcr.bar_chart_race(df = bcr_df, 
                   fixed_order= ['África','Américas', 'Mediterrâneo Oriental', 'Europa', 'Sudeste Asiático','Pacífico Oeste'], 
                   title= 'Novos casos de COVID-19 por região', 
                   perpendicular_bar_func='mean',
                  )
```
